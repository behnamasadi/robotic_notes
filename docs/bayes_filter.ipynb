# Bayes filter 

$$
\text{1. } P(A|B)=\frac{P(A,B)}{P(B)} \\ P(x)=\int P(x,y)dy =\sum P(x|y)P(y) \\ P(x)=\sum P(x,y_i) \\ 
$$

---

$$
\text{2. }P(A|B,C)=\frac{P(A,B,C)} {P(B,C)}=\frac{P(B|A,C)P(A,C)}{P(B|C)P(C)}=\frac{P(B|A,C)P(A|C)P(C)}{P(B|C)P(C)}=\frac{P(B|A,C)P(A|C)}{P(B|C)}
$$

---

$$
\text{3. }P(A,B)=\sum P(A,B,C_i)=\sum P(A|B,C_i)P(B,C_i)=\sum P(A|B,C_i)P(C_i|B)P(B)
$$

---

$$
\text{4. } P(A|B)=\frac{P(A,B)}{P(B)}=\frac{\sum P(A|B,C_i)P(C_i|B)P(B)}{P(B)} =\sum P(A|B,C_i)P(C_i|B)
$$
