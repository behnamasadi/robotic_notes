# Nonlinear uncertainty model associated with a robot's position over time (Banana Shape)
The "banana shape" that forms during state estimation in Simultaneous Localization and Mapping (SLAM) refers to the nonlinear uncertainty model associated with a robot's position over time, especially when turning. This phenomenon is observed in the context of probabilistic SLAM, where the robot's state (position and orientation) and the map of the environment are estimated simultaneously, and uncertainties in these estimates are modeled probabilistically.

### Origin of the Banana Shape

The banana shape arises due to the nonlinear nature of motion and observation models in SLAM. When a robot moves and turns, the uncertainty in its position and orientation does not spread out uniformly. Instead, due to the nonlinearities in the motion (e.g., turning leads to more uncertainty in the direction of the turn than in the perpendicular direction) and observation models (e.g., sensor measurements are more uncertain over longer distances), the probability distribution describing the robot's estimated state becomes skewed and elongated, forming what looks like a banana-shaped curve in the state space.

### Why It's Important

Understanding and effectively dealing with this nonlinear uncertainty is crucial for accurate state estimation in SLAM. It affects how sensor measurements are integrated into the map and the robot's state estimate. Traditional linear estimation techniques, such as the Kalman filter, may not be adequate to capture such complex uncertainty models, leading to the development of nonlinear estimation methods like the Extended Kalman Filter (EKF), the Unscented Kalman Filter (UKF), and particle filters (also known as Monte Carlo Localization).

### Nonlinear Estimation Techniques

1. **Extended Kalman Filter (EKF):** The EKF linearizes the nonlinear models around the current estimate, allowing for the approximation of the banana-shaped distribution. However, it can only approximate the shape to a certain extent and might not be suitable for highly nonlinear systems.

2. **Unscented Kalman Filter (UKF):** The UKF uses a deterministic sampling technique to capture the mean and covariance of the probability distribution more accurately than the EKF, offering a better approximation of the banana shape.

3. **Particle Filter:** The particle filter represents the probability distribution using a set of random samples (particles) and is capable of representing arbitrary distributions, including the banana shape, very well. It is more flexible and accurate in representing nonlinear uncertainties but can be computationally more intensive.

In summary, the banana shape in SLAM state estimation illustrates the complex, nonlinear uncertainties involved in robot localization and mapping. Addressing these uncertainties requires sophisticated nonlinear estimation techniques that can accurately represent and update the robot's state and the map of the environment as new information becomes available.

To illustrate the "banana shape" in the context of nonlinear uncertainty models associated with a robot's position, let's consider a simple example involving a robot moving forward and then turning. The uncertainty in the robot's position increases nonlinearly with motion, especially when incorporating a turn, due to the dependence of the robot's new position on its orientation.

### Equation for Nonlinear Motion Model

A basic nonlinear motion model for a robot moving in a 2D plane can be expressed as follows, where $x_t$ and $y_t$ represent the robot's position at time $t$, and $\theta_t$ its orientation. The robot moves forward by a distance $d$ and then turns by an angle $\alpha$.

$x_{t+1} = x_t + d \cdot \cos(\theta_t)$

$y_{t+1} = y_t + d \cdot \sin(\theta_t)$

$\theta_{t+1} = \theta_t + \alpha$


This model simplifies the robot's motion to a sequence of straight movements and turns, ignoring dynamics and other complexities.

### Illustrating Nonlinear Uncertainty with Python

This code simulates the robot's movement forward and then turning, with randomness in the distance and angle to represent uncertainty. The plot will show the distribution of final positions, illustrating the "banana shape" that arises from the nonlinear relationship between orientation and position in the robot's movement. Let's execute this code to generate the plot.

The plot above illustrates the distribution of final positions for a simulated robot that moves forward and then turns, under the influence of randomness in both the distance moved and the turn angle. This distribution showcases the "banana shape" that emerges due to the nonlinear uncertainty in the robot's motion model. The shape arises because the uncertainty in the robot's final position is not uniform; it's more spread out along the direction of the turn, reflecting the compounded effects of uncertainty in both distance traveled and orientation change. This visualization effectively demonstrates the concept of nonlinear uncertainty in the context of robotic motion and SLAM.

```python
# Re-running simulation with adjusted motion to illustrate "banana shape"
import numpy as np
import matplotlib.pyplot as plt

# Parameters for two movements
num_samples = 2000
d1_mean = 10  # Mean distance moved before the turn
d1_std = 1  # Standard deviation in distance before the turn
d2_mean = 10  # Mean distance moved after the turn
d2_std = 1  # Standard deviation in distance after the turn
alpha_mean = np.pi / 4  # Mean turn angle (45 degrees)
alpha_std = np.pi / 18  # Standard deviation in turn angle (10 degrees)

# Initial state
x0, y0, theta0 = 0, 0, 0  # Starting at the origin

# Simulate first movement
np.random.seed(42)
d1_samples = np.random.normal(d1_mean, d1_std, num_samples)
theta1_samples = theta0 + np.random.normal(alpha_mean, alpha_std, num_samples)

# Calculate intermediate positions after first move
x1 = x0 + d1_samples * np.cos(theta0)
y1 = y0 + d1_samples * np.sin(theta0)

# Simulate second movement after turn
d2_samples = np.random.normal(d2_mean, d2_std, num_samples)
x_final = x1 + d2_samples * np.cos(theta1_samples)
y_final = y1 + d2_samples * np.sin(theta1_samples)

# Plot
plt.figure(figsize=(8, 8))
plt.scatter(x_final, y_final, alpha=0.5)
plt.title('Nonlinear Uncertainty Model: "Banana Shape"')
plt.xlabel('X position')
plt.ylabel('Y position')
plt.grid(True)
plt.axis('equal')
plt.xlim(0, max(x_final) + 10)
plt.ylim(0, max(y_final) + 10)
plt.show()
```
